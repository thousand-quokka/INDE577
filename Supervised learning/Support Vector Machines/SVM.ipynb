# Support Vector Machines (SVM)

In this notebook, we will use SVM (Support Vector Machines) algorithm to classify a dataset of images. SVM is a powerful classification algorithm that can be used in various applications such as image classification, text classification, and bioinformatics.


## Dataset

The dataset we will be using is the MNIST dataset, which consists of 60,000 training images and 10,000 testing images of handwritten digits from 0 to 9. The images are grayscale, with a size of 28 x 28 pixels.

## Model Creation and Training

We will be using scikit-learn's implementation of SVM to create and train our model. We will use the radial basis function (RBF) kernel to map our input data to a higher-dimensional space, which makes it easier to classify.

Here is the code to create and train the SVM model:

```python
from sklearn import datasets, svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# Load the MNIST dataset
mnist = datasets.load_digits()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(mnist.data, mnist.target, test_size=0.2, random_state=42)

# Create the SVM model
model = svm.SVC(kernel='rbf')

# Train the model on the training data
model.fit(X_train, y_train)

# Test the model on the testing data
y_pred = model.predict(X_test)

# Print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

## Model Analysis

Now that we have trained our SVM model, let's analyze its performance. We will visualize the confusion matrix to see which digits the model is having difficulty classifying.

```python
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.matshow(conf_matrix, cmap="magma")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

We can also visualize some of the misclassified images to get a better understanding of where the model is going wrong.

```python
# Find the misclassified images
misclassified_idx = np.where(y_pred != y_test)[0]

# Plot the misclassified images
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(np.reshape(X_test[misclassified_idx[i]], (8, 8)), cmap="magma")
    ax.set_title(f"Predicted: {y_pred[misclassified_idx[i]]}\nActual: {y_test[misclassified_idx[i]]}")
plt.show()
```
